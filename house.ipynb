```python
y=df["MEDV"]
df=df.drop("MEDV",axis=1)
```

```python
from sklearn.preprocessing import RobustScaler,PowerTransformer #zn nox,Dis tax lstat rad
import numpy as np #indus age b pt ratio crim
col =["INDUS","AGE","B","PTRATIO","CRIM"]
for i in col:
       df[i]=np.log1p(df[i])


from sklearn.compose import make_column_transformer
transformer = make_column_transformer(
    (RobustScaler(),["ZN","NOX"]),
    (PowerTransformer(),["DIS","TAX","LSTAT","RAD"]),
    remainder="passthrough"
    )
x=transformer.fit_transform(df)
print(x.shape)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor

gbr = GradientBoostingRegressor()
model=LinearRegression(n_jobs=-1)
ridge=Ridge(alpha=1)
lasso=Lasso(alpha=1)
forest=RandomForestRegressor(random_state=7,n_jobs=-1)

ridge.fit(x_train,y_train)
lasso.fit(x_train,y_train)
model.fit(x_train,y_train)
gbr.fit(x_train, y_train)
forest.fit(x_train,y_train)

ridge_pred=ridge.predict(x_test)
lasso_pred=lasso.predict(x_test)
pred=model.predict(x_test)
forest_pred=forest.predict(x_test)
gbr_pred=gbr.predict(x_test)

from sklearn.metrics import mean_squared_error,r2_score

print(mean_squared_error(pred,y_test))
print(r2_score(pred,y_test))

print(mean_squared_error(lasso_pred,y_test))
print(r2_score(lasso_pred,y_test))

print(mean_squared_error(ridge_pred,y_test))
print(r2_score(ridge_pred,y_test))

print(mean_squared_error(forest_pred,y_test))
print(r2_score(forest_pred,y_test))

print(mean_squared_error(gbr_pred,y_test))
print(r2_score(gbr_pred,y_test))
```
