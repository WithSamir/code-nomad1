```python
x=df["v2"]
y=df["v1"]
```

```python
y.head()
```

```python
from nltk.stem import WordNetLemmatizer
lematizer=WordNetLemmatizer()
def stem(text):
    word=text.split(" ")
    lematize=[lematizer.lemmatize(words) for words in word]
    return " ".join(lematize)
x=x.apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorize=TfidfVectorizer(min_df= 10 or 20,stop_words="english")
x=vectorize.fit_transform(x).toarray()

```

```python
y=y.map({"ham":0,"spam":1})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

from catboost import CatBoostClassifier
model= CatBoostClassifier(task_type="GPU",devices="0",learning_rate=0.05,max_depth=7)
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model2=MultinomialNB()
model2.fit(x_train,y_train)
pred2=model2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,classification_report
print(accuracy_score(y_test,pred))
print(accuracy_score(y_test,pred2))
print(precision_score(y_test,pred))
print(precision_score(y_test,pred2))
print(classification_report(y_test,pred))
print(classification_report(y_test,pred2))
```
