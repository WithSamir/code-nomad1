```python
dropping=["homepage","production_companies","production_countries","runtime","spoken_languages","status","tagline","title","vote_count"]
for i in dropping:
    movies=movies.drop([i],axis=1)
```

```python
movies=movies.dropna()
print(movies.isnull().sum())
print(movies.dtypes)
```

```python
credits["id"]=credits["movie_id"]
credits=credits.drop("movie_id",axis=1)

df=pd.merge(movies,credits,on="id")
df=df.drop("title",axis=1)
df.head()
```

```python
from ast import literal_eval

def extract_names(text):
    try:
        items =literal_eval(text)
        return " ".join([item["name"] for item in items if "name" in item])
    except:
        return ""

for col in ["genres", "keywords", "cast", "crew"]:
    df[col] = df[col].fillna("").apply(extract_names)



df.head()
```

```python
df["crew"]=df["crew"][:3]
df["cast"]=df["cast"][:3]
df.head()
import datetime  as dt

df["release_date"]=pd.to_datetime(df["release_date"])
df["release_date"]=df["release_date"].dt.year
df.head()
df["profit"]=df["revenue"]-df["budget"]
df=((df.drop("budget",axis=1)).drop("revenue",axis=1)).drop("vote_average",axis=1)
df=df.fillna("")
df.head()
```

```python
print(df.columns)
concat=['genres', 'keywords', 'original_language','overview', 'release_date', 'cast', 'crew']
print(df.dtypes)

df["tags"]=""
for i in concat:
    df["tags"]+="" +df[i].astype(str)

for i in concat:
    df=df.drop(i,axis=1)
df["tags"]=df["tags"].str.lower()
df["tags"].iloc[0]
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def convert(text):
    words = text.split(" ")
    lemmatized = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(lemmatized)


df["tags"] = df["tags"].apply(convert)

print(df["tags"][0])

```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

tv=TfidfVectorizer(stop_words="english")
vector=tv.fit_transform(df["tags"]).toarray()
vector
```

```python
df
```

```python
tv.get_feature_names_out()[:500]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vector)

def recommend(rec):
    recommendation=df[df["original_title"]==rec].index[0]
    distance=similar[recommendation]
    list_jnd=(sorted(list(enumerate(distance)),reverse=True,key=lambda x: x[1]))[1:10]
    recommended_indices = [i[0] for i in list_jnd]

   
    newlist=df.iloc[recommended_indices]
    newlist=newlist.sort_values(by="popularity",ascending=False)
    for i in newlist["original_title"]:
        print(i)
recommend("Newlyweds")
```
